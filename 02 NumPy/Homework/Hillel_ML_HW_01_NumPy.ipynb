### Dataset - https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data

```python
# doing necessary imports

import numpy as np
print(np.__version__)
```

```python
# read and save the original dataset

original_ds =np.genfromtxt('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', 
                 dtype='object', delimiter=',', usecols=())
original_ds
```

### 1. Знайти в датасеті таргет та видалити цю колонку з датасету (видаляти за індексом)

```python
# make a copy of original dataset and cut off the last column
ds_ = np.array(original_ds[:, :-1])

# other way - delete the last column
ds = np.delete(original_ds, -1, axis=1)

# ensure both methods make the same result
print(np.array_equal(ds_, ds))

# convert ds to 'float32' type
ds = ds.astype('float32')
ds
```

### 2. Перетворити колонки, що залишились в 2D масив (або впевнитись, що це уже 2D масив)

```python
# ensure we have a 2d array and view other characteristics 
print(ds.ndim, ds.shape, ds.dtype, ds.itemsize, ds.size, sep='\n') 
```

### 3. Порахувати mean, median, standard deviation для 1-ї колонки

```python
# make a view of the first column 
first_col = ds[:, 0]

# calc and print necessary funcs
print(
    np.mean(first_col),
    np.median(first_col),
    np.std(first_col),
    sep='\n'
)
```

### 4. Вставити 20 значень np.nan на випадкові позиції в масиві (при використанні звичайного рандому можуть накластись позиції, тому знайти рішення, яке гарантує 20 унікальних позицій)

```python
np.random.seed(42)

# generate 20 unique values from all elements indexes in ds 
random_20_indexes = np.random.choice(ds.size, size=20, replace=False)

# make a flat view of ds, so we can apply an index array we've got above 
ds_flat_view = ds.ravel()

# set nan to elements in ds applying index array
ds_flat_view[random_20_indexes] = np.nan

print(random_20_indexes)
print(ds)
```

### 6. Відфільтрувати массив за умовою: значення в 3-й колонці > 1.5 та значения в 1-й колонці < 5.0 (зберегти у іншу змінну)

```python
# prepare mask
first_col_mask = ds[:, 0] < 5
third_col_mask = ds[:, 2] > 1.5
mask = first_col_mask & third_col_mask

# apply mask to ds
filtered_ds = ds[mask]
print(filtered_ds)
```

### 7. Замінити всі значення np.nan на 0

```python
# prepare mask
nan_mask = np.isnan(ds)

# apply mask
ds = np.where(nan_mask, 0, ds)
print(ds)
```

### 8. Порахувати всі унікальні значення в массиві та вивести їх разом із кількістю 

```python
values, counts = np.unique(ds, return_counts=True)
print(values, counts, sep='\n')
```

### 9. Розбити масив по вертикалі на 2 рівні частини (не використовувати абсолютні числа, мають бути два массиви по 4 колонки)

```python
upper_ds, lower_ds = np.vsplit(ds, 2)
print(upper_ds, lower_ds, sep='\n')
```

### 10. Відсортувати обидва массиви по 1-й колонці: 1-й за збільшенням, 2-й за зменшенням

```python
# prepare sorting masks
upper_ds_sort_index = upper_ds[:, 0].argsort()
lower_ds_sort_index = lower_ds[:, 0].argsort()[::-1]  # [::-1] inverts order 

# doing sort by applying sorting masks
upper_ds = upper_ds[upper_ds_sort_index]
lower_ds = lower_ds[lower_ds_sort_index]
print(upper_ds, lower_ds, sep='\n')
```

### 11. Зібрати обидва массиви в одне ціле

```python
ds_new = np.vstack([upper_ds, lower_ds])
print(ds_new)
```

### 12. Знайти найбільш часто повторюване значення в массиві

```python
values, counts = np.unique(ds_new, return_counts=True)
print(values, counts, sep='\n')

most_frequently_value = values[counts == counts.max()][0]
print(most_frequently_value)
```

### 13. Написати функцію, яка б множила всі значення в колонці, які менше середнього значения в цій колонці, на 2, і ділила інші значення на 4.

```python
def my_func(col: np.ndarray) -> np.ndarray:
    average = col.mean()
    col = np.where(col < average, col * 2, col / 4)
    return col
```

### 14. Застосувати отриману функцію до 3-ї колонки

```python
# the third column is ds_new[:,2]
ds_new[:,2] = my_func(ds_new[:,2])
print(ds_new)
```
